## Exercise 1 • Test scores
### Create a series of 10 elements, random integers from 70 to 100, representing scores on a monthly exam. Set the index to be the month names, starting in September and ending in June. (If these months don’t match the school year in your location, feel free to make them more realistic.)

### With this series, write code to answer the following questions:
### • What is the student’s average test score for the entire year?
### • What is the student’s average test score during the first half of the year (i.e., the first five months)?
### • What is the student’s average test score during the second half of the year?
### • Did the student improve their performance in the second half? If so, by how much?

```python
import pandas as pd, numpy as np

gen = np.random.default_rng(16580)
scores = pd.Series(gen.integers(70, 101, 10), index = ['September', 'October', 'November', 'December', 'January', 'February', 'March', 'April', 'May', 'June'])
```

```python
scores
```

```python
print(f"The average of the student's test scores is {float(scores.mean())}.")

sem1, sem2 = float(scores[:5].mean()), float(scores[5:].mean())

print(f"The average of the student's test scores in the first 5 months is {sem1}.")
print(f"The average of the student's test scores in the last 5 months is {sem2}.")

print(f"{f'An improvement of {sem2 - sem1:.3f}% was made.' if sem1 < sem2 else f'A deterioration of {sem1 - sem2:.3f}% was made.' if sem1 > sem2 else 'Neither an improvement or deterioration was made.'}")
```

```python
print(f'The student acheived the highest score in the month of {scores.idxmax()}.')
```
