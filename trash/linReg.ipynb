$MSE(\mathbf{X}, h_{\boldsymbol{\theta}})=\frac{1}{m}\sum^{m}_{i=1}(\boldsymbol{\theta}^T\mathbf{x}^{(i)}-y^{(i)})^2\\$
$\mathbf{X}$ is the training set. It is a vector of all $\mathbf{x}$, where $\mathbf{x}$ is a vector containing all the feature values of a specific datapoint.$\\$
$m$ is the number of datapoints.$\\$
$\boldsymbol{\theta}$ is the weights/hyperperameters.$\\$
$y$ is the predicted value.$\\$
$\theta_0$ is always 1 to represent bias.

```python
import numpy as np
m = 100 # number of instances
X = 2 * np.random.rand(m, 1) # column vector
y = 4 + 3 * X + np.random.randn(m, 1) # column vector
X.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color="blue", alpha=0.6, label="data points")

# plot the true line without noise
x_line = np.linspace(0, 2, 100)
y_line = 4 + 3 * x_line
plt.plot(x_line, y_line, color="red", linewidth=2, label="true function")

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
error = y - (4 + 3 * X)
plt.hist(error, bins=20)
```

```python
new = np.ones((m,1))
X2 = np.hstack((new, X))
X2[0:5]
```

```python
w = np.zeros((2,1))
lr = 0.1
for epoch in range(1000):
    w -= lr * 2/m * X2.T @ (X2 @ w - y)
w
```
